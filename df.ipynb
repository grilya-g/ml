```python
accountancy = pd.read_csv('accountancy.csv')
```

```python
accountancy
```

```python
mean_salary_per_type = accountancy.groupby(['Executor', 'Type'])\
    .aggregate({'Salary' : 'mean'})
mean_salary_per_type
```

```python
mean_salary_per_type.loc[('Pupa', 'D')][0]
```

```python
for t in accountancy['Type'].unique():
    if mean_salary_per_type.loc[('Pupa', t)][0] > mean_salary_per_type.loc[('Loopa', t)][0]:
        print(t, "Pupa")
    else:
        print(t, "Loopa")
    # print(mean_salary_per_type.loc[('Pupa', t)])
```
